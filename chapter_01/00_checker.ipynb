```python
# Design Pattern
"""
객체지향 디자인 패턴
- 객체지향 프로그래밍을 할 때 자주 발생하는 문제들을 피하기 위해 사용하는 패턴

디자인 패턴의 종류
- 생성 패턴
    - 객체의 생성과 관련된 패턴
    - 객체의 생성과 조합을 캡슐화하여 객체를 생성하는 패턴
    - 생성 패턴의 예
        - Singleton
        - Factory
- 구조 패턴
    - 클래스나 객체를 조합하여 더 큰 구조를 만드는 패턴
    - 구조 패턴의 예
        - Facade
        - Proxy
- 행위 패턴
    - 클래스나 객체가 상호작용하는 방법과 책임을 분산하는 패턴
    - 행위 패턴의 예
        - State
        - Command
        - Observer
        
객체
    - 속성
    - 함수(메소드)

상속과 컴포지션
    - 상속
        - 부모 클래스의 속성과 함수를 그대로 물려 받는 것
        - 부모 클래스의 속성과 함수를 재정의(Overriding)하여 사용 가능
    - 컴포지션
        - 다른 클래스의 객체를 속성으로 사용하는 것
        - 다른 클래스의 객체를 속성으로 사용하여 함수를 호출
        - 다른 클래스의 객체를 속성으로 사용하여 함수를 호출하고, 그 결과를 반환

객체지향 디자인 원칙

- 단일 책임 원칙(Single Responsibility Principle)
    - 하나의 클래스는 하나의 책임만 가져야 한다.

- 개방-폐쇄 원칙(Open-Closed Principle)
    - 확장에는 열려 있어야 하고, 변경에는 닫혀 있어야 한다.
    - 기존의 코드를 변경하지 않으면서 기능을 추가할 수 있도록 설계가 되어야 한다.
    - 다형성을 이용하여 구현한다.

- 리스코프 치환 원칙(Liskov Substitution Principle)
    - 자식 클래스는 최소한 자신의 부모 클래스에서 가능한 행위는 수행할 수 있어야 한다.
    - 다형성을 지원하기 위한 원칙이다.

- 인터페이스 분리 원칙(Interface Segregation Principle)
    - 자신이 사용하지 않는 인터페이스는 구현하지 말아야 한다.
    - 자신이 사용하지 않는 기능에는 종속되지 않도록 설계해야 한다.

- 의존 역전 원칙(Dependency Inversion Principle)
    - 자신보다 변하기 쉬운 것에 의존하지 말아야 한다.
    - 구체적인 클래스보다 인터페이스나 추상 클래스와 관계를 맺도록 설계해야 한다.
    - 다형성을 지원하기 위한 원칙이다.

- 컴포지션/집합 원칙(Composition/Aggregation Principle)
    - 클래스의 객체들을 속성으로 가지는 것을 의미한다.
    - 상속은 is-a 관계를 표현하는데 사용하고, 컴포지션은 has-a 관계를 표현하는데 사용한다.

- 디미터 법칙(Law of Demeter)
    - 자신이 조작하는 객체의 메소드만 호출한다.
    - 메소드에서 생성한 객체의 메소드는 호출하지 않는다.
    - 매개변수로 받은 객체의 메소드는 호출하지 않는다.
    - 자신이 생성한 객체의 메소드만 호출한다.

"""
```
